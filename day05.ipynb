```python
from collections import defaultdict
from functools import cmp_to_key

def get_input():
    order, lines = open('inputs/day05.txt').read().split('\n\n')

    order = order.split('\n')
    lines = lines.split('\n')

    required = defaultdict(list)
    for line in order: 
        needed, result = line.split("|")
        required[result].append(needed)
    return required, lines


# Custom sort function to sort based on requirements
def custom_sort(x, y): 
    if x in required[y]: 
        return -1
    if y in required[x]: 
        return 1
    return 0
    
required, lines = get_input()
sum_middle_part1 = 0
sum_middle_part2 = 0

for line in lines: 
    if len(line) == 0: 
        continue
    line = line.split(",")

    is_correct = True
    
    # Make sure that all numbers are in the right order
    for i in range(len(line)): 
        if line[i] in required: 
            for req in required[line[i]]: 
                if req in line[i:]: 
                    is_correct = False
                    break
    if is_correct: 
        sum_middle_part1 += int(line[(len(line))//2])
    else: 
        sorted_line = sorted(line, key=cmp_to_key(custom_sort))
        sum_middle_part2 += int(sorted_line[(len(sorted_line))//2])
    
print('Part 1:', sum_middle_part1)
print('Part 2:', sum_middle_part2)
```
