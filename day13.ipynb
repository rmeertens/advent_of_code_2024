```python
import re
import sympy

lines = open('inputs/day13.txt').read()
arcade_machines = [x for x in lines.split('\n\n') if len(x) > 5]

print('number of machines', len(arcade_machines))
# ['Button A: X+40, Y+38\nButton B: X+21, Y+84\nPrize: X=4245, Y=5634',


def parse_arcade_machine(machine_string): 
    ax, ay, bx, by, prizex, prizey = [int(x) for x in re.findall(r'\d+', machine_string)]
    return ax, ay, bx, by, prizex, prizey

def steps_to_prize(ax, ay, bx, by, prizex, prizey): 
    # Find count_a and count_b such that ax*count_a + bx*count_b = prizex and ay*count_a + by*count_b = prizey
    # Using sympy
    count_a, count_b = sympy.symbols('count_a count_b')
    eq1 = sympy.Eq(ax*count_a + bx*count_b, prizex)
    eq2 = sympy.Eq(ay*count_a + by*count_b, prizey)
    solution = sympy.solve([eq1, eq2], (count_a, count_b))
    if not solution: 
        return -1
    print('solution', solution)
    presses_a = solution[count_a]
    presses_b = solution[count_b]

    # Check if presses_a and presses_b are a whole number
    if presses_a.is_integer and presses_b.is_integer:
        return 3*int(presses_a) + 1*int(presses_b)
    else: 
        return -1
    return -1
    # return 3*solution[count_a] + 1*solution[count_b]



    # max_presses = 10000
    # for a_press in range(max_presses): 
    #     for b_press in range(max_presses): 
    #         if ax*a_press + bx*b_press == prizex and ay*a_press + by*b_press == prizey: 
    #             print('found', a_press, b_press)
    #             return 3*a_press + 1*b_press
            
    # return -1
            
total_tokens = 0           
part_2 = True 
for machine in arcade_machines: 
    ax, ay, bx, by, prizex, prizey = parse_arcade_machine(machine)

    if part_2: 
        prizex, prizey = 10000000000000 + prizex, 10000000000000 + prizey

    presses = steps_to_prize(ax, ay, bx, by, prizex, prizey)
    print(machine)
    print(presses)
    if presses != -1: 
        
        total_tokens += presses
print('total tokens', total_tokens)        
```
