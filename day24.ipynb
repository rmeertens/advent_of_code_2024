```python
from collections import defaultdict

file = open('inputs/day24.txt').read()
initial_set, connections = file.split('\n\n')

initial_settings = dict()
connect_dict = dict()
for line in initial_set.split('\n'):
    register, number = line.split(': ')
    initial_settings[register] = int(number)
    connect_dict[register] = int(number)


for line in connections.split('\n'):
    if len(line) > 0:
        reg1, operation, reg2, _, result_reg = line.split()
        connect_dict[result_reg] = (reg1, operation, reg2)

    
z_values = sorted([key for key in connect_dict.keys() if key[0] == 'z'], reverse=True)

def get_value(register, connect_dict):
    if type(connect_dict[register]) == int:
        return connect_dict[register]
    else:
        reg1, operation, reg2 = connect_dict[register]
        if operation == 'AND':
            return get_value(reg1, connect_dict) & get_value(reg2, connect_dict)
        elif operation == 'OR':
            return get_value(reg1, connect_dict) | get_value(reg2, connect_dict)
        elif operation == 'XOR':
            return get_value(reg1, connect_dict) ^ get_value(reg2, connect_dict)
        else:
            raise ValueError('Unknown operation')

binary_number = ""
for z in z_values:
    binary_number += str(get_value(z, connect_dict))
print('Part 1', int(binary_number, 2))

# Part 2
# Print everything in dot code to visualize the circuit
if visualise_circuit:
    counter = defaultdict(int)
    for a in connect_dict: 
        if type(connect_dict[a]) != int:
            in1, op, in2 = connect_dict[a]
            print(f"{in1} -> {op}_{counter[op]}")
            print(f"{in2} -> {op}_{counter[op]}")
            print(f"{op}_{counter[op]} -> {a}")

            counter[op] += 1

new_connect_dict = connect_dict.copy()
new_connect_dict['gws'], new_connect_dict['nnt'] = new_connect_dict['nnt'], new_connect_dict['gws']
new_connect_dict['z33'], new_connect_dict['hgj'] = new_connect_dict['hgj'], new_connect_dict['z33']
new_connect_dict['z19'], new_connect_dict['cph'] = new_connect_dict['cph'], new_connect_dict['z19']
new_connect_dict['z13'], new_connect_dict['npf'] = new_connect_dict['npf'], new_connect_dict['z13']        


def get_number(connect_dict):
    number_str = ""
    for z in sorted(z_values, reverse=True):
        number_str += str(get_value(z, connect_dict))
    return int(number_str, 2)

def get_x(connect_dict):
    number_str = ""
    for x in sorted(x_values, reverse=True):
        number_str += str(get_value(x, connect_dict))
    return int(number_str, 2)

def get_y(connect_dict):
    number_str = ""
    for y in sorted(y_values, reverse=True):
        number_str += str(get_value(y, connect_dict))
    return int(number_str, 2)

x_values = [key for key in connect_dict.keys() if key[0] == 'x']
y_values = [key for key in connect_dict.keys() if key[0] == 'y']

for z in z_values[1:-1]: 
    integer_value = int(z[1:])
    for cur_y in range(2): 
        for cur_x in range(2): 
            for prev_xy in range(2): 
                # new_initial_settings = defaultdict(int)
                prev_x_register = "x{:02}".format(integer_value-1)
                prev_y_register = "y{:02}".format(integer_value-1)
                cur_x_register = "x{:02}".format(integer_value)
                cur_y_register = "y{:02}".format(integer_value)

                # Set all xy values to 0
                for x in x_values: 
                    new_connect_dict[x] = 0
                for y in y_values:
                    new_connect_dict[y] = 0

                new_connect_dict[prev_x_register] = prev_xy
                new_connect_dict[prev_y_register] = prev_xy
                new_connect_dict[cur_x_register] = cur_x
                new_connect_dict[cur_y_register] = cur_y

                z_value = get_value(z, new_connect_dict)
                # print(z, z_value)
                # print(get_x(connect_dict), get_y(connect_dict), get_number(connect_dict))
                x_val = get_x(new_connect_dict)
                y_val = get_y(new_connect_dict)
                z_val = get_number(new_connect_dict)
                if x_val + y_val != z_val:
                    print("Error", x_val, y_val, z_val, z)
                
print('Part 2', ','.join(sorted(['gws', 'nnt', 'z33', 'hgj', 'z19', 'cph', 'z13', 'npf'])))
```
